```python
from selenium import webdriver
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
def get_content_and_related_articles(row):
    global driver
    article_content=''
    related_content=''
    img_links=''

    url=row['link']
    # Edge浏览器
    driver.get(url)
    print("retrieved {}/{}th url".format(row.name,df.shape[0]))

    #找文章内容图片html
    #图片img class="rich_pages wxw-img" data-src
    wait = WebDriverWait(driver, 10)
    try:
        try:
            content_imgs = wait.until(EC.visibility_of_all_elements_located((By.CSS_SELECTOR, '.rich_pages.wxw-img')))  
             #找到它们的链接
            img_links = [i.get_attribute('data-src') for i in content_imgs]
            print(img_links)
        except:
            wait = WebDriverWait(driver, 10)
            content_imgs=wait.until(EC.presence_of_all_elements_located((By.CSS_SELECTOR,'img[data-src]')))
            img_links = [i.get_attribute('data-src') for i in content_imgs]
            print(img_links)
        #找相关文章推荐html
        related_divs = wait.until(EC.visibility_of_all_elements_located((By.CSS_SELECTOR,'div[id^="relate_article_title"]')))
        related_content = [i.text for i in related_divs]
        print(related_content)

        #找文章内容html 
        content_divs=wait.until(EC.visibility_of_all_elements_located((By.ID,'page-content')))
        article_content = [i.text for i in content_divs]
    except:
        print("{}/n has been deleted by the author".format(url))
        return 'The article has been deleted by the author','The article has been deleted by the author','The article has been deleted by the author'
    
    return article_content,related_content,img_links

```
